# 4. 数据特征处理

## 4.1 便利设施  
对数据集中的“便利设施”这一列进行处理，变成新的一列“便利设施_num”——里面的信息为原列里面设施数量

```python
train['便利设施_num'] = train['便利设施'].str.split(',')
int_equipments = []
for item in train['便利设施_num']:
    num = len(item)
    int_equipments.append(num)
train['便利设施_num'] = int_equipments


test['便利设施_num'] = test['便利设施'].str.split(',')
int_equipments_test = []
for item in test['便利设施_num']:
    num = len(item)
    int_equipments_test.append(num)
test['便利设施_num'] = int_equipments_test
```

为特定关键词创建新的二元列，并将其转换成字符串格式，以便后续使用LabelEncoder进行统一处理

```python
for key_words in ['Internet', 'Kitchen', 'Pets', 'Air', 'Washer', 'Aid', 'Heating', 'Dryer', 'Essentials', 'Hangers', 'TV', 'parking', '24-hour']:
    train[key_words] = train['便利设施'].str.contains(key_words).astype('str')
    test[key_words] = test['便利设施'].str.contains(key_words).astype('str')
```

## 4.2 缺失值处理  

### 4.2.1民宿周边，使用-999进行缺失值填充，之后缺失值记为0，非缺失值记为1

```python
train['民宿周边'] = train['民宿周边'].fillna(-999)
test['民宿周边'] = test['民宿周边'].fillna(-999)
surround = []
for value in train['民宿周边']:
    if value == -999:
        new_value = 0
    else:
        new_value = 1
    surround.append(new_value)
train['民宿周边'] = surround

surround_test = []
for value in test['民宿周边']:
    if value == -999:
        new_value = 0
    else:
        new_value = 1
    surround_test.append(new_value)
test['民宿周边'] = surround_test
```

针对某一些列，使用 f 来进行缺失值填充  
对另外的某一些列，需要进行Label编码，将字符串值转换成整数，以便用于机器模型的训练，对于每列，使用fit_transform方法在训练集上拟合编码器并对训练集和测试集进行编码

```python
train['房主是否有个人资料图片'] = train['房主是否有个人资料图片'].fillna('f')
test['房主是否有个人资料图片'] = test['房主是否有个人资料图片'].fillna('f')

surround1 = []
for value in train['房主是否有个人资料图片']:
    if value == 'f':
        new_value = 0
    else:
        new_value = 1
    surround1.append(new_value)
train['房主是否有个人资料图片'] = surround1

surround1_test = []
for value in test['房主是否有个人资料图片']:
    if value == 'f':
        new_value = 0
    else:
        new_value = 1
    surround1_test.append(new_value)
test['房主是否有个人资料图片'] = surround1_test
```

```python
train['房主身份是否验证'] = train['房主身份是否验证'].fillna('f')
test['房主身份是否验证'] = test['房主身份是否验证'].fillna('f')

surround2 = []
for value in train['房主身份是否验证']:
    if value == 'f':
        new_value = 0
    else:
        new_value = 1
    surround2.append(new_value)
train['房主身份是否验证'] = surround2

surround2_test = []
for value in test['房主身份是否验证']:
    if value == 'f':
        new_value = 0
    else:
        new_value = 1
    surround2_test.append(new_value)
test['房主身份是否验证'] = surround2_test
```

```python
str_cols = ['Internet', 'Kitchen', 'Pets', 'Air', 'Washer', 'Aid', 'Heating', 'Dryer', 'Essentials', 'Hangers', 'TV', 'parking', '24-hour']

le = LabelEncoder()
for col in str_cols:
    train[col] = le.fit_transform(train[col])
    test[col] = le.fit_transform(test[col])
```

### 4.2.2 房主回复率  
将缺失值填充为20%（随意取的），再去掉最后的%字符，这样就将百分比值转换成浮点数值。

```python
train['房主回复率'] = train['房主回复率'].fillna('20%')
test['房主回复率'] = test['房主回复率'].fillna('20%')
train['房主回复率'] = train['房主回复率'].str[:-1]
test['房主回复率'] = test['房主回复率'].str[:-1]
```

### 4.2.3 日期数据  
使用2019-1-1填充缺失值（随意选的），将日期时间列转换成日期时间数据类型，以便后续可以在日期时间特征中使用

```python
train['首次评论日期'] = train['首次评论日期'].fillna('2019-1-1')
test['首次评论日期'] = test['首次评论日期'].fillna('2019-1-1')
train['最近评论日期'] = train['最近评论日期'].fillna('2019-1-1')
test['最近评论日期'] = test['最近评论日期'].fillna('2019-1-1')

time_cols = ['首次评论日期', '最近评论日期', '何时成为房主']
for col in time_cols:
    train[col] = pd.to_datetime(train[col])
    test[col] = pd.to_datetime(test[col])
```

### 4.2.4 将数据集划分成连续变量和分类变量，分别进行不同的处理  
将分类变量数据类型转换成category，以便后续处理  
连续变量：中位数  
分类变量：众数  
最后重置索引，以确保数据集中不再存在缺失值

```python
cols = train.columns
cont_cols = ['容纳人数', '便利设施_num', '床的数量', '卧室数量', '洗手间数量', '维度', '经度', '评论个数', '民宿评分', '房主回复率']
cate_cols = ['Internet', 'Kitchen', 'Air', 'Pets', 'Washer', 'Aid', 'Heating', 'Dryer', 'Essentials', 'Hangers',
        'TV', 'parking', '24-hour', '床的类型', '民宿周边', '取消条款', '所在城市', '清洁费', '是否支持随即预订', 
        '房产类型', '房型', '房主是否有个人资料图片', '房主身份是否验证']
features = cont_cols + cate_cols
train[cate_cols] = train[cate_cols].astype('category')
test[cate_cols] = test[cate_cols].astype('category')

for col in cate_cols:
    train[col] = train[col].fillna(train[col].mode())
    test[col] = test[col].fillna(test[col].mode())

for col in ['洗手间数量', '床的数量', '卧室数量', '民宿评分']:
    train[col] = train[col].fillna(train[col].median())
    test[col] = test[col].fillna(test[col].median())

train = train.dropna().reset_index()
```

### 4.2.5 标准化处理（对连续变量进行特征缩放）  
标准差标准化：均值为0，标准差为1的分布  
四分位数范围标准化：对数据中的离群值不敏感  
Min-Max标准化：通常缩放到0-1之间  
使用fit方法对连续变量进行拟合，之后使用transform()方法进行特征缩放

```python
X = train[features]

scaler = RobustScaler()
# scaler = StandardScaler()
# scaler = MinMaxScaler()

scaler.fit(X[cont_cols])

X[cont_cols] = scaler.transform(X[cont_cols])

X_test = test[features]
X_test[cont_cols] = scaler.transform(X_test[cont_cols])

y = train['价格'].astype('int')
```
